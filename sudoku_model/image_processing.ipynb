```python
img = cv2.imread('sudoku.png')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
```

```python
img_resize = cv2.resize(img, (26*9, 26*9))
```

```python
img_resize.shape
```

```python
plt.imshow(img_resize)
```

```python
edge  = cv2.Canny(img, 100, 100 )
plt.imshow(edge)
```

```python
contours, hierarchy = cv2.findContours(edge, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE) 
```

```python
contours
```

```python
plt.imshow(cv2.drawContours(edge, contours, 1, (0,255,255), 1))
```

```python
len(contours)
```

```python
max = 0

for i in contours:
    if(cv2.contourArea(i)>max):
        max = cv2.contourArea(i)
```
